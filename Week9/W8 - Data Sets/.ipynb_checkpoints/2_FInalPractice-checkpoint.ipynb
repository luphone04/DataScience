# RUNNING A MODEL

```python
from statsmodels.formula.api import ols

#creates a linear regression model
# specifies that you want to model the 'total_payment_sek' variable as a linear function of the 'n_claims' variable
# dependent variable / Response variable ('total_payment_sek') and 
#the independent variable / Explanatory variable('n_claims')
mdl_payment_vs_claims = ols("total_payment_sek ~ n_claims", data = swedish_motor_insurance)

#calculates the coefficients (parameters) of the model 
#that best describe the relationship between 'total_payment_sek' and 'n_claims'.
mdl_payment_vs_claims = mdl_payment_vs_claims.fit()

# prints the intercept and the coefficient for the 
#'n_claims' variable. These coefficients represent the 
#intercept and the slope of the regression line
print(mdl_payment_vs_claims.params)
```

```python
#Interpreting the model coefficients
#Equation
#total_payment_sek = 19.99 + 3.14 * n_claims
```

# .resid attribute

"actual response" values minus "predicted response values"

RESIDUALS: ACTUAL RESPONSE VALUES MINUS PREDICTED RESPONSES VALUES

"Residuals" are a measure of inaccuracy in the model fit, and are accessed with the .resid attribute. Like fitted values, there is one residual for each row of the dataset. Each residual is the actual response value minus the predicted response value. In this case, the residuals are the masses of breams, minus the fitted values. Here we illustrated the residuals as red lines on the regression plot. Each vertical line represents a single residual.

```python
print(mdl_mass_vs_length.resid)
```

```python
print(bream["mass_g"] - mdl_mass_vs_length.fittedvalues)
```

Manually predicting house prices

You can manually calculate the predictions from the model coefficients. When making predictions in real life, it is better to use .predict(), but doing this manually is helpful to reassure yourself that predictions aren't magic - they are simply arithmetic.

In fact, for a simple linear regression, the predicted value is just the intercept plus the slope times the explanatory variable.

Here, explanatory_data is a number of convenience store from 0 to10.

 Get the coefficients/parameters of mdl_price_vs_conv, assigning to coeffs.

 Get the intercept, which is the first element of coeffs, assigning to intercept.

 Get the slope, which is the second element of coeffs, assigning to slope.

 Manually predict price_twd_msq using the formula, specifying the intercept, slope, and explanatory_data.

 Run the code to compare your manually calculated predictions to the results from .predict().

```python
print(mdl_price_vs_conv.params)
```

```python
explanatory_data = pd.DataFrame({'n_convenience': np.arange(0,11)})

#coefficient/ parameters
coeffs = mdl_price_vs_conv.params
intercept = coeffs['Intercept']
slope = coeffs['n_convenience']
price_twd_msq = intercept + slope * explanatory_data
print(price_twd_msq)
print(price_twd_msq.assign(predictions_auto=mdl_price_vs_conv.predict(explanatory_data)))
```

sns.regplot is primarily used for creating scatterplots with a regression line (hence the "reg" in the name).
It not only displays the individual data points but also fits a regression line to represent the overall trend or relationship between the variables.
The regression line helps you understand the linear relationship between the two variables and provides insight into how one variable changes as the other variable changes.
It's useful when you want to visualize and quantify the linear association between variables.
sns.scatterplot:

sns.scatterplot is more focused on displaying individual data points as they are, without fitting a regression line.
It is versatile and can be used for general scatterplot purposes, including comparing individual data points, identifying clusters or patterns, and visualizing relationships without imposing a specific model (such as linear regression).
It provides more flexibility in customizing the appearance of data points, such as marker style, size, and color.

In summary, the key difference is that sns.regplot emphasizes the linear relationship with a regression line, while sns.scatterplot is more general and suitable for displaying individual data points and making comparisons without fitting a specific model. # QUANTIFYING MODEL FIT

HOW GOOD OR BAD THE MODEL IS

# .SUMMARY()

```python
mdl_bream = ols("mass_g ~ length_cm", data = bream).fit()
print(mdl_bream.summary())
```

The summary you've posted is the output of a linear regression analysis performed using the Ordinary Least Squares (OLS) method. It provides various statistics and information about the regression model. Here's how to interpret the key parts of this summary:

Dependent Variable: The "Dep. Variable" section indicates the variable you are trying to predict. In your case, it's "mass_g," which represents the mass of something in grams.

Model Information:

"Model": This shows that you used the OLS method for linear regression.
"Method": This indicates the method used to estimate the model parameters.
"Date" and "Time": The date and time when the analysis was performed.
R-squared and Adj. R-squared:

"R-squared" (R-squared or R^2): This measures the goodness-of-fit of the model. It represents the proportion of variance in the dependent variable (mass_g) that is explained by the independent variable (length_cm). In this case, approximately 87.8% of the variance in mass_g is explained by the linear relationship with length_cm.
"Adj. R-squared" (Adjusted R-squared): This is a modified version of R-squared that adjusts for the number of predictors in the model. It helps prevent overfitting. An adjusted R-squared of 0.874 means that the model explains a significant amount of variance in mass_g while considering model complexity.
F-statistic and Prob (F-statistic):

"F-statistic": This tests the overall significance of the model. In this case, it's 237.6, and a higher value suggests that at least one independent variable significantly contributes to explaining the variance in the dependent variable.
"Prob (F-statistic)": This is the p-value associated with the F-statistic. In this case, the p-value is very close to zero (1.22e-16), indicating that the model is statistically significant.

# .RSQUARED ATTRIBUTE

```python
print(mdl_bream.rsquared)
```

R-squared = 0: None of the variance in the dependent variable is explained by the independent variable(s). The model does not fit the data at all.

R-squared = 1: The independent variable(s) perfectly explain(s) the variance in the dependent variable. The model fits the data perfectly.

0 < R-squared < 1: The R-squared value indicates the proportion of variance explained by the model. For example, an R-squared value of 0.878, as you mentioned in your previous output, means that approximately 87.8% of the variance in "mass_g" is explained by the linear relationship with "length_cm."

# .MSE_RESID ATTRIBUTE

represents the Mean Squared Error (MSE) of the model's residuals

The residuals are the differences between the observed values of the dependent variable ("mass_g" in this case) and the predicted values generated by the linear regression model.

a measure of the average squared difference between the observed and predicted values. It quantifies the overall goodness of fit of the model to the data, with lower values indicating a better fit. 

```python
mse = mdl_bream.mse_resid
print('mse: ', mse)
```

```python
rse = np.sqrt(mse)
print("rse: ", rse)
```
