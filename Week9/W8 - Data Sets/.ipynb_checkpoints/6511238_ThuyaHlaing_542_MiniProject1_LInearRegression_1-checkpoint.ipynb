```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
ad_conversion = pd.read_csv("ad_conversion.csv")
sns.regplot(x="n_impressions", y="n_clicks", data=ad_conversion,ci=None)
plt.show()
```

```python
ad_conversion["qdrt_n_impressions"] = ad_conversion["n_impressions"]**0.25
ad_conversion["qdrt_n_clicks"] = ad_conversion["n_clicks"]**0.25
sns.regplot(x="qdrt_n_impressions", y="qdrt_n_clicks", data=ad_conversion, ci=None)
plt.show()
```

```python
from statsmodels.formula.api import ols
mdl_click_vs_impression = ols("qdrt_n_clicks ~ qdrt_n_impressions", data=ad_conversion).fit()
mdl_click_vs_impression.params
```

```python
explanatroy_data = pd.DataFrame({"qdrt_n_impressions": np.arange(0, 3000000, 500000)**0.25, "n_impressions": np.arange(0, 3000000, 500000)})
prediction_data = explanatroy_data.assign(
    qdrt_n_clicks = mdl_click_vs_impression.predict(explanatroy_data))
print(prediction_data)
```

```python
fig = plt.figure()
sns.regplot(x = "qdrt_n_impressions", y="qdrt_n_clicks", data=ad_conversion, ci=None)
sns.scatterplot(data = prediction_data, x="qdrt_n_impressions", y="qdrt_n_clicks", color="red", marker="s")
```

```python
prediction_data['n_clicks'] = prediction_data['qdrt_n_clicks']**4
sns.regplot(x='n_impressions', y='n_clicks', data=ad_conversion, ci=None)
sns.scatterplot(x='n_impressions', y='n_clicks', data=prediction_data, color='red', marker='s')
```

```python
residuals = ad_conversion['qdrt_n_clicks'] - mdl_click_vs_impression.predict(ad_conversion[['qdrt_n_impressions']])
rse = np.sqrt((residuals ** 2).mean())
print('Residual Standard Error (RSE):', rse)
```
