```python
# Get the top 10 countries
top_countries = sorted_countries.sort_values('title', ascending=False).head(10)

# Calculate the total count of titles
total_titles = top_countries['title'].sum()

# Plotting the outer circle (total count)
outer_labels = ['Total Titles', '']
outer_sizes = [total_titles, 0]
outer_colors = ['#D3D3D3', 'white']
plt.pie(outer_sizes, labels=outer_labels, colors=outer_colors, wedgeprops={'edgecolor': 'white'})

# Plotting the inner circle (count of titles by country)
inner_labels = top_countries.index
inner_sizes = top_countries['title']
inner_colors = plt.cm.Set3(np.arange(len(inner_labels)))
plt.pie(inner_sizes, labels=inner_labels, colors=inner_colors, autopct='%1.1f%%', wedgeprops={'edgecolor': 'white'})

# Add a white circle in the middle to create the donut shape
center_circle = plt.Circle((0, 0), 0.65, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

# Setting the title
plt.title('Number of Titles Released by Top 10 Countries')

# Displaying the plot
plt.axis('equal')
plt.show()
```
