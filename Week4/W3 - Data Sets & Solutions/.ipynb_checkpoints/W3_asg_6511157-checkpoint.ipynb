```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
worldbank = pd.read_csv("WorldBank_GDP.csv")

worldbankChina = worldbank[worldbank['Country Name'] == 'China']
worldbankGermany = worldbank[worldbank['Country Name'] == 'Germany']
worldbankJapan = worldbank[worldbank['Country Name'] == 'Japan']
worldbankUS = worldbank[worldbank['Country Name'] == 'United States']

plt.plot(worldbankGermany['Year'], worldbankGermany['GDP'], label='Germany')
plt.plot(worldbankChina['Year'], worldbankChina['GDP'], label='China')
plt.plot(worldbankJapan['Year'], worldbankJapan['GDP'], label='Japan')
plt.plot(worldbankUS['Year'], worldbankUS['GDP'], label='United States')

plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend()
plt.show()
```

```python
import pandas as pd

temperature = pd.read_csv("temperatures.csv")

avg_temp_by_country = temperature.groupby('country')['avg_temp_c'].mean()

highest_avg_temp = avg_temp_by_country.max()


countries_with_highest_temp = avg_temp_by_country[avg_temp_by_country == highest_avg_temp]

for country, avg_temp in countries_with_highest_temp.items():
    print("Country:", country)
    print("Max Average Temperature:", avg_temp)
```
